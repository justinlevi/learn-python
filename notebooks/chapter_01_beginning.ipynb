### ğŸ“š What are Variables?

Think of variables as **labeled boxes** where you can store information. Just like you might have a box labeled "LEGOs" that holds your LEGO bricks, in Python you can have a variable called `robot_name` that holds the name of your robot.

In Python, creating a variable is super easy - you just give it a name and assign it a value using the `=` sign:

```python
variable_name = value
```

### ğŸ¨ Data Types

Python has several types of data. Here are the three most important ones:

1. **String (str)** - Text, always in quotes
   - `"Hello"`, `'Python'`, `"Robotics Club"`
   
2. **Integer (int)** - Whole numbers
   - `42`, `-7`, `0`, `2024`
   
3. **Float** - Decimal numbers
   - `3.14`, `-0.5`, `98.6`

Python automatically figures out what type your variable is!

````python
# Let's create some variables!

# String (text) - notice the quotes
hero_name = "Aragorn"
print(f"Hero name: {hero_name}")
print(f"Type: {type(hero_name)}")
print()

# Integer (whole number) - no quotes!
hero_level = 1
print(f"Hero level: {hero_level}")
print(f"Type: {type(hero_level)}")
print()

# Float (decimal number)
hero_health = 100.0
print(f"Hero health: {hero_health}")
print(f"Type: {type(hero_health)}")
````

### ğŸ’¡ Key Points:

- Variable names should be descriptive: `hero_name` is better than `x`
- Use lowercase with underscores: `player_health` (this is called "snake_case")
- Variable names can't start with numbers: `1st_place` âŒ but `first_place` âœ…
- Python is case-sensitive: `Name` and `name` are different variables!

### ğŸ¨ What are F-Strings?

F-strings (formatted string literals) are the **modern way** to put variables inside strings in Python. They make your code clean and easy to read!

**Syntax:** Put an `f` before the quote, then use `{variable_name}` inside the string.

```python
f"Text {variable} more text"
```

### Why F-Strings?

Compare these approaches:

```python
# âŒ Old way (don't use this)
print("My name is " + name + " and I am " + str(age) + " years old.")

# âœ… Modern way (use this!)
print(f"My name is {name} and I am {age} years old.")
```

F-strings are:
- âœ… Easier to read
- âœ… Easier to write
- âœ… Faster to run
- âœ… Automatically convert numbers to text

````python
# F-String Examples

name = "Gandalf"
age = 2000
location = "The Shire"

# Basic f-string
print(f"My name is {name} and I am {age} years old.")

# You can do math inside f-strings!
print(f"In 100 years, I'll be {age + 100} years old!")

# Multiple variables
print(f"{name} is {age} years old and lives in {location}.")

# You can format numbers
pi = 3.14159265359
print(f"Pi rounded to 2 decimals: {pi:.2f}")  # .2f means 2 decimal places
````

### ğŸ¤ The input() Function

To make your programs interactive, you need to get information from the user. The `input()` function does exactly that!

```python
variable_name = input("Your question here: ")
```

### âš ï¸ Important: input() Always Returns a String!

Even if the user types a number, Python treats it as text. If you want a number, you need to **convert** it:

```python
age_str = input("How old are you? ")  # This is a string: "15"
age = int(age_str)                     # Convert to integer: 15
```

**Shortcut:** You can convert in one line:

```python
age = int(input("How old are you? "))
```

````python
# Example: Getting user input

# Get a string (text)
name = input("What is your name? ")
print(f"Hello, {name}!")
print()

# Get a number - need to convert!
age_str = input("What is your age? ")
age = int(age_str)  # Convert string to integer
print(f"Wow, you're {age} years old!")
print(f"Next year you'll be {age + 1}!")  # Now we can do math!
print()

# Shortcut: convert in one line
level = int(input("What level are you? "))
print(f"Level {level} adventurer! Impressive!")
````

### ğŸ”§ Type Conversion Functions:

- `int()` - Convert to integer (whole number)
- `float()` - Convert to decimal number
- `str()` - Convert to string (text)

```python
int("42")      # Returns: 42
float("3.14")  # Returns: 3.14
str(42)        # Returns: "42"
```

## âœ… TEST SETUP - Run this first!
---

**âš ï¸ IMPORTANT: Run this cell before attempting any challenges!**

This sets up the testing framework that will check your code.

````python
# Test Framework Setup
# Run this cell first!

class TestResult:
    """Helper class to track and display test results"""
    def __init__(self):
        self.passed = []
        self.failed = []
    
    def add_pass(self, test_name):
        self.passed.append(test_name)
        print(f"âœ… PASS: {test_name}")
    
    def add_fail(self, test_name, reason):
        self.failed.append((test_name, reason))
        print(f"âŒ FAIL: {test_name}")
        print(f"   Reason: {reason}")
    
    def summary(self):
        total = len(self.passed) + len(self.failed)
        print("\n" + "â•" * 60)
        print(f"RESULTS: {len(self.passed)}/{total} tests passed")
        print("â•" * 60)
        if len(self.failed) == 0:
            print("ğŸ‰ ALL TESTS PASSED! Great work!")
            print("âœ¨ You're ready for the next challenge!")
        else:
            print("\nâš ï¸  Some tests failed. Review the feedback above.")
            print("ğŸ’¡ TIP: Read the error messages carefully - they tell you what's wrong!")

def test_variable(name, expected_type, test_results, expected_value=None):
    """Helper function to test if a variable exists and has correct type"""
    try:
        # Check if variable exists in global namespace
        if name not in globals():
            test_results.add_fail(
                f"Variable '{name}' exists",
                f"Variable '{name}' not found. Did you create it?"
            )
            return False
        
        actual = globals()[name]
        
        # Check type
        if not isinstance(actual, expected_type):
            test_results.add_fail(
                f"Variable '{name}' has correct type",
                f"Expected {expected_type.__name__}, got {type(actual).__name__}"
            )
            return False
        
        test_results.add_pass(f"Variable '{name}' exists with type {expected_type.__name__}")
        
        # Check value if provided
        if expected_value is not None:
            if actual == expected_value:
                test_results.add_pass(f"Variable '{name}' has correct value")
            else:
                test_results.add_fail(
                    f"Variable '{name}' has correct value",
                    f"Expected {expected_value}, got {actual}"
                )
                return False
        
        return True
    except Exception as e:
        test_results.add_fail(f"Variable '{name}' test", str(e))
        return False

print("âœ… Test framework loaded successfully!")
print("ğŸ“ You can now complete the challenges below.")
print("ğŸ¯ Good luck, adventurer!")
````
