## â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

### ğŸ” What is a While Loop?

A **while loop** repeats code as long as a condition is `True`. It keeps checking the condition before each repetition.

**Syntax:**
```python
while condition:
    # This code repeats while condition is True
    # Don't forget indentation!
```

### ğŸ”‘ Key Points:
1. The condition is checked **before** each iteration
2. If the condition starts as `False`, the loop never runs
3. The loop continues **until** the condition becomes `False`
4. You must **change something** inside the loop to eventually make it stop!

### âš ï¸ Danger: Infinite Loops!

If the condition **never** becomes `False`, you create an **infinite loop**:

```python
# âŒ DON'T DO THIS - Infinite loop!
while True:
    print("Forever!")
    # Nothing changes, so it never stops!
```

**How to stop an infinite loop in Jupyter/Colab:**
- Click the "Stop" button (â¹ï¸)
- Or use Kernel â†’ Interrupt

### âœ… Good While Loop Pattern:

```python
counter = 0
while counter < 5:
    print(f"Count: {counter}")
    counter += 1  # Change the variable to eventually make condition False!
```

````python
# Simple while loop example

count = 1

while count <= 5:
    print(f"Count: {count}")
    count += 1  # Same as: count = count + 1

print("Done!")
print(f"Final count: {count}")
````

````python
# Game example: Health regeneration

hero_health = 50
max_health = 100

print(f"Starting health: {hero_health}/{max_health}")
print("Resting...\n")

turns = 0
while hero_health < max_health:
    hero_health += 10  # Restore 10 HP per turn
    turns += 1
    print(f"Turn {turns}: Health restored to {hero_health}/{max_health}")

print(f"\nFully healed in {turns} turns!")
````

### ğŸ”¢ What is a For Loop?

A **for loop** repeats code a **specific number of times**. It's perfect when you know exactly how many iterations you need.

**Syntax:**
```python
for variable in sequence:
    # This code repeats for each item in the sequence
```

### ğŸ“ The range() Function

`range()` creates a sequence of numbers, perfect for for loops!

**Three ways to use range():**

1. **range(stop)** - From 0 to stop (not including stop)
   ```python
   range(5)  # 0, 1, 2, 3, 4
   ```

2. **range(start, stop)** - From start to stop (not including stop)
   ```python
   range(1, 6)  # 1, 2, 3, 4, 5
   ```

3. **range(start, stop, step)** - With a custom step size
   ```python
   range(0, 10, 2)  # 0, 2, 4, 6, 8
   ```

### ğŸ”‘ Key Points:
- The loop variable automatically gets each value
- `range(5)` gives you 5 numbers: 0, 1, 2, 3, 4
- The stop value is **never included**
- For loops are great when you know the number of iterations

````python
# Basic for loop with range

print("Counting from 0 to 4:")
for i in range(5):
    print(f"Number: {i}")

print("\nCounting from 1 to 5:")
for i in range(1, 6):
    print(f"Number: {i}")

print("\nEven numbers 0 to 10:")
for i in range(0, 11, 2):
    print(f"Number: {i}")
````

````python
# Game example: Attack sequence

print("âš”ï¸ COMBO ATTACK! âš”ï¸\n")

total_damage = 0

for hit in range(1, 6):  # 5 hits: 1, 2, 3, 4, 5
    damage = hit * 10  # Damage increases each hit!
    total_damage += damage
    print(f"Hit {hit}: {damage} damage!")

print(f"\nğŸ’¥ Total damage: {total_damage}!")
````

## â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

### ğŸ›‘ Loop Control: break and continue

Sometimes you need more control over your loops. Python provides two special keywords:

### ğŸ›‘ break - Exit the Loop Immediately

Use `break` to exit a loop early, even if the condition is still True.

```python
while True:
    user_input = input("Continue? (y/n): ")
    if user_input == "n":
        break  # Exit the loop!
    print("Continuing...")
```

### â­ï¸ continue - Skip to Next Iteration

Use `continue` to skip the rest of the current iteration and jump to the next one.

```python
for i in range(5):
    if i == 2:
        continue  # Skip when i is 2
    print(i)  # Prints: 0, 1, 3, 4 (skips 2)
```

### ğŸ”‘ When to Use What:
- **break** - Stop the loop entirely (found what you're looking for, error occurred, user wants to quit)
- **continue** - Skip this iteration but keep looping (invalid data, special case to skip)

### ğŸ’¡ Common Pattern: Infinite Loop with break

```python
while True:  # Infinite loop!
    choice = input("Enter choice (or 'quit'): ")
    if choice == "quit":
        break  # Only way to exit
    # Process the choice
```

````python
# break example: Finding treasure

print("Searching through chests...\n")

for chest in range(1, 11):
    print(f"Opening chest {chest}...")
    
    if chest == 7:
        print("  ğŸ‰ Found the legendary sword!")
        print("  No need to search more!")
        break  # Stop searching!
    else:
        print("  Empty chest.")

print(f"\nStopped at chest {chest}")
````

````python
# continue example: Skip certain numbers

print("Counting, but skipping multiples of 3:\n")

for num in range(1, 11):
    if num % 3 == 0:  # % is modulo (remainder)
        continue  # Skip multiples of 3
    
    print(num)

print("\nDone! (Skipped 3, 6, 9)")
````

````python
# Game example: Combat with break

enemy_health = 100
player_attacks = 0

print("âš”ï¸ BATTLE START! âš”ï¸")
print(f"Enemy Health: {enemy_health}\n")

while True:
    player_attacks += 1
    damage = 30
    enemy_health -= damage
    
    print(f"Attack {player_attacks}: -{damage} damage!")
    print(f"Enemy health: {enemy_health}")
    
    if enemy_health <= 0:
        print("\nğŸ‰ Enemy defeated!")
        break  # Exit the infinite loop
    
    print()  # Blank line

print(f"Victory in {player_attacks} attacks!")
````

## ğŸ¯ INTERACTIVE EXAMPLE 3: Break and Continue
---

````python
# Try this example - it combines break and continue!

print("Searching for healing potions...\n")

potions_found = 0
target_potions = 3

for location in range(1, 11):
    print(f"Searching location {location}...", end=" ")
    
    # Skip odd locations (no items there)
    if location % 2 != 0:
        print("(Empty)")
        continue
    
    # Even locations have potions!
    potions_found += 1
    print(f"âœ¨ Found potion #{potions_found}!")
    
    # Stop when we have enough
    if potions_found >= target_potions:
        print(f"\nCollected {target_potions} potions! That's enough!")
        break

print(f"Search ended at location {location}")
````

## ğŸ® GAME FEATURE DEMO: Main Game Loop
## â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

Now let's put it all together with a complete game loop!

````python
# Complete Game Loop Example

print("â•" * 60)
print("ADVENTURE GAME - EXPLORE THE DUNGEON".center(60))
print("â•" * 60)
print()

# Player stats
health = 100
gold = 0
rooms_explored = 0

# Main game loop
while True:
    print("\n" + "-" * 60)
    print(f"HP: {health} | Gold: {gold} | Rooms explored: {rooms_explored}")
    print("-" * 60)
    
    # Check for game over
    if health <= 0:
        print("\nğŸ’€ You have died! Game Over!")
        break
    
    # Show menu
    print("\nWhat would you like to do?")
    print("1. Explore a room")
    print("2. Rest (restore 20 HP)")
    print("3. Check stats")
    print("4. Quit game")
    
    choice = input("\nYour choice: ")
    
    if choice == "1":
        # Explore a room
        rooms_explored += 1
        print(f"\nğŸšª Entering room {rooms_explored}...")
        
        # Random event (simplified with modulo)
        event = rooms_explored % 3
        
        if event == 0:
            print("ğŸ’° You found a treasure chest!")
            found_gold = 25
            gold += found_gold
            print(f"Gained {found_gold} gold!")
        elif event == 1:
            print("âš”ï¸ A monster attacks!")
            damage = 20
            health -= damage
            print(f"You took {damage} damage!")
        else:
            print("ğŸ•¸ï¸ The room is empty and dusty.")
    
    elif choice == "2":
        # Rest
        print("\nğŸ˜´ You rest for a while...")
        heal = 20
        health = min(health + heal, 100)  # Don't exceed max health
        print(f"Restored {heal} HP!")
    
    elif choice == "3":
        # Stats
        print("\nğŸ“Š YOUR STATS:")
        print(f"   Health: {health}/100")
        print(f"   Gold: {gold}")
        print(f"   Rooms explored: {rooms_explored}")
    
    elif choice == "4":
        # Quit
        print("\nğŸ‘‹ Thanks for playing!")
        break
    
    else:
        print("\nâŒ Invalid choice! Try again.")
        continue

# Game over - show final stats
print("\n" + "â•" * 60)
print("FINAL STATS".center(60))
print("â•" * 60)
print(f"Rooms explored: {rooms_explored}")
print(f"Gold collected: {gold}")
print(f"Final health: {health}")
print("â•" * 60)
````

````python
# Test Framework Setup

class TestResult:
    def __init__(self):
        self.passed = []
        self.failed = []
    
    def add_pass(self, test_name):
        self.passed.append(test_name)
        print(f"âœ… PASS: {test_name}")
    
    def add_fail(self, test_name, reason):
        self.failed.append((test_name, reason))
        print(f"âŒ FAIL: {test_name}")
        print(f"   Reason: {reason}")
    
    def summary(self):
        total = len(self.passed) + len(self.failed)
        print("\n" + "â•" * 60)
        print(f"RESULTS: {len(self.passed)}/{total} tests passed")
        print("â•" * 60)
        if len(self.failed) == 0:
            print("ğŸ‰ ALL TESTS PASSED! Great work!")
        else:
            print("\nâš ï¸  Some tests failed. Review the feedback above.")

print("âœ… Test framework loaded!")
print("ğŸ¯ Ready for loop challenges!")
````
