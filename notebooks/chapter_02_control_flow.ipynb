### ğŸ”µ What are Booleans?

A **boolean** is a data type that can only be one of two values:
- `True` (yes, correct, 1)
- `False` (no, incorrect, 0)

Booleans are named after George Boole, a mathematician who invented Boolean algebra!

**Important:** In Python, `True` and `False` must be capitalized!

### âš–ï¸ Comparison Operators

Comparison operators let you compare values and get a boolean result:

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `==` | Equal to | `5 == 5` | `True` |
| `!=` | Not equal to | `5 != 3` | `True` |
| `<` | Less than | `3 < 5` | `True` |
| `>` | Greater than | `5 > 3` | `True` |
| `<=` | Less than or equal | `5 <= 5` | `True` |
| `>=` | Greater than or equal | `5 >= 3` | `True` |

### âš ï¸ Common Mistake:
- `=` is for **assignment** (creating variables)
- `==` is for **comparison** (checking if values are equal)

```python
x = 5      # Assignment: x is now 5
x == 5     # Comparison: is x equal to 5? Returns True
```

````python
# Boolean values and comparison examples

# Direct boolean values
is_hero = True
is_enemy = False
print(f"Is hero: {is_hero}")
print(f"Is enemy: {is_enemy}")
print()

# Comparison operators
hero_level = 5
enemy_level = 3

print(f"Hero level: {hero_level}")
print(f"Enemy level: {enemy_level}")
print()

print(f"Hero level == 5: {hero_level == 5}")  # True
print(f"Hero level != 3: {hero_level != 3}")  # True
print(f"Hero level > enemy level: {hero_level > enemy_level}")  # True
print(f"Hero level < enemy level: {hero_level < enemy_level}")  # False
print(f"Hero level >= 5: {hero_level >= 5}")  # True
print()

# Comparing strings
hero_class = "Warrior"
print(f"Hero class == 'Warrior': {hero_class == 'Warrior'}")  # True
print(f"Hero class == 'Mage': {hero_class == 'Mage'}")  # False
````

### ğŸ”€ Making Decisions with if

The `if` statement lets your program make decisions. It runs code **only if** a condition is True.

**Syntax:**
```python
if condition:
    # This code runs if condition is True
    # Note the indentation (4 spaces)!
```

### ğŸ”‘ Key Points:
1. The condition must result in a boolean (True/False)
2. Don't forget the colon `:` at the end of the if line!
3. The code inside must be **indented** (use 4 spaces or Tab)
4. Multiple lines can be inside the if block - all must be indented the same

### ğŸ“ Indentation is CRUCIAL!

Python uses indentation to know what code belongs to the if statement. Other languages use `{}` braces, but Python uses spaces!

```python
if hero_health > 0:
    print("Hero is alive!")      # Indented - runs if True
    print("Keep fighting!")       # Indented - runs if True
print("Game continues")           # Not indented - always runs
```

### ğŸ”€ Multiple Paths: elif and else

Often you want to check multiple conditions, or have a "default" action. That's where `elif` and `else` come in!

**Syntax:**
```python
if condition1:
    # Runs if condition1 is True
elif condition2:
    # Runs if condition1 is False AND condition2 is True
elif condition3:
    # Runs if condition1 and condition2 are False AND condition3 is True
else:
    # Runs if ALL conditions above are False
```

### ğŸ”‘ Key Points:
- `elif` is short for "else if"
- You can have as many `elif` blocks as you want
- `else` is optional and always comes last
- **Only ONE block executes!** Once a condition is True, Python skips the rest

### ğŸ’¡ When to use what:
- **if** - Check a condition
- **elif** - Check another condition if the first was False
- **else** - Default action if nothing else was True

````python
# if/elif/else example

hero_health = 75

if hero_health >= 80:
    print("ğŸ’š Health: Excellent")
    print("You're in peak condition!")
elif hero_health >= 50:
    print("ğŸ’› Health: Good")
    print("You're doing well.")
elif hero_health >= 20:
    print("ğŸ§¡ Health: Low")
    print("Consider healing!")
else:
    print("â¤ï¸ Health: Critical!")
    print("Heal immediately!")

print(f"\nCurrent HP: {hero_health}")
````

### ğŸ”— Combining Conditions

Sometimes you need to check **multiple conditions** at once. Logical operators let you combine boolean values!

### The Three Logical Operators:

1. **`and`** - Both conditions must be True
   ```python
   if age >= 13 and age <= 19:
       print("You're a teenager")
   ```

2. **`or`** - At least one condition must be True
   ```python
   if day == "Saturday" or day == "Sunday":
       print("It's the weekend!")
   ```

3. **`not`** - Reverses the boolean value
   ```python
   if not is_game_over:
       print("Game continues!")
   ```

### ğŸ“Š Truth Tables:

**AND** - Both must be True:
| A | B | A and B |
|---|---|------|
| True | True | True |
| True | False | False |
| False | True | False |
| False | False | False |

**OR** - At least one must be True:
| A | B | A or B |
|---|---|------|
| True | True | True |
| True | False | True |
| False | True | True |
| False | False | False |

**NOT** - Reverses the value:
| A | not A |
|---|------|
| True | False |
| False | True |

````python
# Logical operator examples

hero_level = 5
hero_gold = 75
has_key = True

# AND example - both conditions must be True
if hero_level >= 5 and hero_gold >= 50:
    print("âœ… You can enter the advanced dungeon!")
    print("   (Requires level 5+ AND 50+ gold)")
print()

# OR example - at least one condition must be True
if hero_level >= 10 or has_key:
    print("âœ… You can open the locked door!")
    print("   (Requires level 10+ OR a key)")
print()

# NOT example - reverses the boolean
is_game_over = False
if not is_game_over:
    print("âœ… Game is still running!")
print()

# Complex combination
if (hero_level >= 5 and hero_gold >= 50) or has_key:
    print("âœ… You can access the treasure room!")
    print("   (Requires (level 5+ AND 50+ gold) OR a key)")
````

Here's a complete example of a branching story system using everything we've learned!

````python
# Complete Branching Story System

print("â•" * 60)
print("THE FOREST PATH".center(60))
print("â•" * 60)
print()

# Get player info
hero_name = input("Enter your hero's name: ")
hero_class = input("Enter your class (Warrior/Mage/Rogue): ")

print(f"\nWelcome, {hero_name} the {hero_class}!\n")

# Starting stats
health = 100
gold = 50

# Story begins
print("You're walking through a dark forest when you hear a noise...")
print("A wolf jumps out in front of you!")
print()
print("What do you do?")
print("1. Attack with your weapon")
print("2. Try to tame it")
print("3. Run away")

choice = input("\nYour choice (1-3): ")
print()

if choice == "1":
    print("âš”ï¸ You attack the wolf!")
    
    if hero_class == "Warrior":
        print("Your warrior training pays off!")
        print("You defeat the wolf easily.")
        gold += 20
        print(f"You gained 20 gold! Total: {gold}")
    elif hero_class == "Mage":
        print("You cast a spell, but it takes effort.")
        print("You defeat the wolf but use some health.")
        health -= 10
        gold += 15
        print(f"HP: {health}, Gold: {gold}")
    else:  # Rogue
        print("You strike from the shadows!")
        print("Critical hit! The wolf runs away.")
        gold += 25
        print(f"You found treasure! Gold: {gold}")

elif choice == "2":
    print("ğŸº You try to calm the wolf...")
    
    if hero_class == "Rogue" or hero_class == "Mage":
        print("Your gentle approach works!")
        print("The wolf becomes your companion!")
        print("ğŸ You gained: Wolf companion")
    else:
        print("The wolf doesn't trust you...")
        print("It bites you before running away!")
        health -= 20
        print(f"HP: {health}")

elif choice == "3":
    print("ğŸƒ You run away!")
    health -= 5
    print("You escape but trip and hurt yourself.")
    print(f"HP: {health}")

else:
    print("âŒ You hesitate!")
    print("The wolf attacks!")
    health -= 30
    print(f"HP: {health}")

print()
print("â•" * 60)
print("ADVENTURE CONTINUES...")
print(f"Final Stats - HP: {health}, Gold: {gold}")
print("â•" * 60)
````

````python
# Test Framework Setup
# Run this cell first!

class TestResult:
    def __init__(self):
        self.passed = []
        self.failed = []
    
    def add_pass(self, test_name):
        self.passed.append(test_name)
        print(f"âœ… PASS: {test_name}")
    
    def add_fail(self, test_name, reason):
        self.failed.append((test_name, reason))
        print(f"âŒ FAIL: {test_name}")
        print(f"   Reason: {reason}")
    
    def summary(self):
        total = len(self.passed) + len(self.failed)
        print("\n" + "â•" * 60)
        print(f"RESULTS: {len(self.passed)}/{total} tests passed")
        print("â•" * 60)
        if len(self.failed) == 0:
            print("ğŸ‰ ALL TESTS PASSED! Great work!")
        else:
            print("\nâš ï¸  Some tests failed. Review the feedback above.")

print("âœ… Test framework loaded!")
print("ğŸ¯ Ready for challenges!")
````
